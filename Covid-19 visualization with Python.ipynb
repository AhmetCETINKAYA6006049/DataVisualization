```python
import numpy as np
import pandas as pd
```

```python
covid_data = pd.read_csv('owid-covid-data.csv')
```

```python
covid_data.head(5)
```

https://www.listendata.com/2019/07/how-to-filter-pandas-dataframe.html 

filter only the countries we want

```python
turkey_covid_data = covid_data[covid_data.location == "Turkey"]
turkey_covid_data.head(5)
```

https://realpython.com/pandas-sort-python/#getting-familiar-with-sort_index

```python
turkey_covid_data.sort_values("date")
```

show daily statistics in Turkey  last 15 days

```python
last_15_days_covid = turkey_covid_data[turkey_covid_data.date > '2021-07-10']
last_15_days_covid.head(15)
```

plot daily cases as barplot 
https://queirozf.com/entries/pandas-dataframe-plot-examples-with-matplotlib-pyplot

```python
last_15_days_covid.plot(kind = "bar", x= "date", y = "new_cases")
```

plot total deaths and new deaths last 15 days as line plot

```python
import matplotlib.pyplot as plt
ax = plt.gca()
last_15_days_covid.plot(kind = "line", x= "date", y = "total_deaths", ax = ax, xlabel = "date", ylabel ="count")
last_15_days_covid.plot(kind = "line", x= "date", y = "new_deaths", color = 'red', ax = ax, xlabel = "date", ylabel ="count" )
plt.yscale("log")
plt.xticks(rotation = 90)
```
