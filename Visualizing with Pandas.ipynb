# Visualizing with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The NumPy function randn() generates a NumPy array with a
standard normal random distribution. We will use this to create a series
and plot it

```python
ts = pd.Series(np.random.randn(5))
ts
```

```python
ts.plot()
plt.grid(True)
plt.show()
```

We can compute the cumulative sum with the function cumsum() and
visualize it

```python
ts.cumsum().plot()
plt.grid(True)
plt.show()
```

Now letâ€™s create a dataframe as follows

```python
df1 = pd.DataFrame(np.random.randn(10, 4),
columns=['A', 'B', 'C', 'D'])
print(df1)
```

We can visualize this with vertical bar graphs

```python
df1.plot.bar()
plt.grid(True)
plt.show()
```

use horizontal bar graphs

```python
df1.plot.barh()
plt.grid(True)
plt.show()
```

stacked vertical bar graphs

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B','C', 'D'])
print(df2)
```

```python
df2.plot.bar(stacked=True)
plt.grid(True)
plt.show()
```

horizontal stacked bar plots

```python
df2.plot.barh(stacked=True)
plt.grid(True)
plt.show()
```

an area plot

```python
df2.plot.area(stacked=True)
plt.grid(True)
plt.show()
```

```python
df2.plot.area(stacked=False)
plt.grid(True)
plt.show()
```

a pie chart

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
